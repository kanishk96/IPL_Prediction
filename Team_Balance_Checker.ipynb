```python
import pandas as pd 
  
# intialise data of lists. 
data = {'Teams':['Chennai Super Kings','Delhi Capitals','Kolkata Knight Riders','Kings XI Punjab','Mumbai Indians','Rajasthan Royals','Royal Challengers Bangalore','Sunrisers Hyderabad'],'Batsman':[1,2,3,4,5,6,7,8],'Allrounders':[1,2,3,4,5,6,7,8],'Bowlers':[1,2,3,4,5,6,7,8]} 
  
# Create DataFrame 
df = pd.DataFrame(data) 

CSK_df=pd.read_csv("Chennai_Super_Kings.csv")
DC_df=pd.read_csv("Delhi_Capitals.csv")
KKR_df=pd.read_csv("KKR.csv")
KXIP_df=pd.read_csv("KXIP.csv")
MI_df=pd.read_csv("Mumbai_Indians.csv")
RR_df=pd.read_csv("Rajasthan_Royals.csv")
RCB_df=pd.read_csv("Royal_Challengers_Bangalore.csv")
SRH_df=pd.read_csv("SRH.csv")

# Print the output. 
df=df.set_index('Teams')

#function to set categories
def set_category(row):
    if (row['Role'] == 'Opening batsman' or row['Role'] == 'Batsman' or row['Role'] == 'Opening Wicketkeeper batsman' or row['Role'] == 'Wicketkeeper batsman'):
        return 'Batsman'
    if (row['Role'] == 'Bowler' or row['Role'] == 'Spin bowler'):
        return 'Bowler'
    if (row['Role'] == 'Allrounder' or row['Role'] == 'All round spin bowler'):
        return 'Allrounder'

#Chennai    
CSK_df['Category'] = CSK_df.apply (lambda row: set_category(row), axis=1)
CSK_df['Category'].value_counts()
CSK_df = CSK_df['Category'].value_counts().reset_index()
CSK_df.columns = ['Category', 'count']
df['Batsman'][0]=CSK_df['count'][1]
df['Allrounders'][0]=CSK_df['count'][2]
df['Bowlers'][0]=CSK_df['count'][0]

#Delhi
DC_df['Category'] = DC_df.apply (lambda row: set_category(row), axis=1)
DC_df['Category'].value_counts()
DC_df = DC_df['Category'].value_counts().reset_index()
DC_df.columns = ['Category', 'count']
df['Batsman'][1]=DC_df['count'][1]
df['Allrounders'][1]=DC_df['count'][2]
df['Bowlers'][1]=DC_df['count'][0]


#KKR
KKR_df['Category'] = KKR_df.apply (lambda row: set_category(row), axis=1)
KKR_df['Category'].value_counts()
KKR_df = KKR_df['Category'].value_counts().reset_index()
KKR_df.columns = ['Category', 'count']
df['Batsman'][2]=KKR_df['count'][1]
df['Allrounders'][2]=KKR_df['count'][2]
df['Bowlers'][2]=KKR_df['count'][0]

#KXIP
KXIP_df['Category'] = KXIP_df.apply (lambda row: set_category(row), axis=1)
KXIP_df['Category'].value_counts()
KXIP_df = KXIP_df['Category'].value_counts().reset_index()
KXIP_df.columns = ['Category', 'count']
df['Batsman'][3]=KXIP_df['count'][1]
df['Allrounders'][3]=KXIP_df['count'][2]
df['Bowlers'][3]=KXIP_df['count'][0]

#MI
MI_df['Category'] = MI_df.apply (lambda row: set_category(row), axis=1)
MI_df['Category'].value_counts()
MI_df = MI_df['Category'].value_counts().reset_index()
MI_df.columns = ['Category', 'count']
df['Batsman'][4]=MI_df['count'][1]
df['Allrounders'][4]=MI_df['count'][2]
df['Bowlers'][4]=MI_df['count'][0]

#RR
RR_df['Category'] = RR_df.apply (lambda row: set_category(row), axis=1)
RR_df['Category'].value_counts()
RR_df = RR_df['Category'].value_counts().reset_index()
RR_df.columns = ['Category', 'count']
df['Batsman'][5]=RR_df['count'][1]
df['Allrounders'][5]=RR_df['count'][2]
df['Bowlers'][5]=RR_df['count'][0]

#RCB
RCB_df['Category'] = RCB_df.apply (lambda row: set_category(row), axis=1)
RCB_df['Category'].value_counts()
RCB_df = RCB_df['Category'].value_counts().reset_index()
RCB_df.columns = ['Category', 'count']
df['Batsman'][6]=RCB_df['count'][1]
df['Allrounders'][6]=RCB_df['count'][2]
df['Bowlers'][6]=RCB_df['count'][0]

#SRH
SRH_df['Category'] = SRH_df.apply (lambda row: set_category(row), axis=1)
SRH_df['Category'].value_counts()
SRH_df = SRH_df['Category'].value_counts().reset_index()
SRH_df.columns = ['Category', 'count']
df['Batsman'][7]=SRH_df['count'][1]
df['Allrounders'][7]=SRH_df['count'][2]
df['Bowlers'][7]=SRH_df['count'][0]
df
```
