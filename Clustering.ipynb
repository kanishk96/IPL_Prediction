```python
import numpy as np 
import pandas as pd
from scipy import ndimage 
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 
from matplotlib import pyplot as plt 
from sklearn import manifold, datasets 
from sklearn.cluster import AgglomerativeClustering 
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline


CSK_df=pd.read_csv("Chennai_Super_Kings.csv")
CSK_df['Name']=CSK_df['Name'].map(lambda x: str(x)[:-12])
CSK_df['Average'][CSK_df['Average'] == '-'] = 0
CSK_df['Average'] = CSK_df['Average'].astype(float)
CSK_df['Strike Rate'][CSK_df['Strike Rate'] == '-'] = 0
CSK_df['Strike Rate'] = CSK_df['Strike Rate'].astype(float)
CSK_df['Wickets'][CSK_df['Wickets'] == '-'] = 0
CSK_df['Wickets'] = CSK_df['Wickets'].astype(float)
CSK_df['Economy'][CSK_df['Economy'] == '-'] = 0
CSK_df['Economy'] = CSK_df['Economy'].astype(float)
CSK_df['Bowling Average'][CSK_df['Bowling Average'] == '-'] = 0
CSK_df['Bowling Average'] = CSK_df['Bowling Average'].astype(float)
CSK_df['Bowling Strike Rate'][CSK_df['Bowling Strike Rate'] == '-'] = 0
CSK_df['Bowling Strike Rate'] = CSK_df['Bowling Strike Rate'].astype(float)

#featureset
featureset = CSK_df[['Matches',  'Runs', 'Average', 'Strike Rate', 'Wickets', 'Bowling Average', 'Bowling Strike Rate', 'Economy']]

#normalization

from sklearn.preprocessing import MinMaxScaler
x = featureset.values #returns a numpy array
min_max_scaler = MinMaxScaler()
feature_mtx = min_max_scaler.fit_transform(x)
feature_mtx [0:5]

#clusters
agglom = AgglomerativeClustering(n_clusters = 3, linkage = 'complete')
agglom.fit(feature_mtx)
agglom.labels_

#displaying the cluster number
CSK_df['Cluster'] = agglom.labels_
CSK_df

#plotting the cluster
import matplotlib.cm as cm
n_clusters = max(agglom.labels_)+1
colors = cm.rainbow(np.linspace(0, 1, n_clusters))
cluster_labels = list(range(0, n_clusters))

# Create a figure of size 6 inches by 4 inches.
plt.figure(figsize=(16,14))

for color, label in zip(colors, cluster_labels):
    subset = CSK_df[CSK_df.Cluster == label]
    for i in subset.index:
            plt.text(subset.Runs[i], subset.Wickets[i],str(subset['Name'][i]), rotation=25) 
    plt.scatter(subset.Runs, subset.Wickets, s= subset.Matches, c=color, label='Cluster'+str(label),alpha=0.5)
#    plt.scatter(subset.horsepow, subset.mpg)
plt.legend()
plt.title('Clusters')
plt.xlabel('Runs')
plt.ylabel('Wickets')
```
